# Pearson相關係數

### 思路： 
### 1. 列出每個商品每天的銷量,在兩兩做相關分析
### 2.以當日銷量佔總體銷量比例來做為變數,不以實際銷量數做變數

```python
import csv
lines = []
with open('tag_date_rate.csv', 'r') as f:
    for line in csv.reader(f):
        lines.append(line)   #此檔案內容為:品名,日期,當日銷量佔總體銷量比例

lines
```

```python
import csv
tags = []
with open('tag_distinct.csv', 'r') as f:
    for line in f:
        tags.append(line.strip())
del tags[0]

tags
```

```python
from datetime import datetime, timedelta
start = "2013-12-31"
end = "2017-06-15"
s = datetime.strptime(start, "%Y-%m-%d")
e = datetime.strptime(end, "%Y-%m-%d")
dates = []   #做出時間順序tags
while s <= e:
    sf = s.strftime("%Y-%m-%d")
    dates.append(sf)
    s += timedelta(1)
    
dates
```

```python
import pandas as pd
df = pd.DataFrame([list([0.0001]*len(tags))]*len(dates),index=dates,columns=tags)   #建立一張datas()*tags,0.0001的dataframe
with open('tag_date_rate.csv', 'r') as f:
    for line in csv.reader(f):
        try:
            df.set_value(line[1],line[0],float(line[2]))   #將line[2]的值放進對應的row:line[1](time), column:line[0](tag)
        except ValueError:
            df.set_value(line[1],line[0],0)
df.drop(df.index[1097:])    #頭尾資料怪怪的,去掉
df.drop(df.index[0])
```

```python
#.iloc可輕易的選取dataframe內的某欄某值,ex:
#.iloc[0:1,1]  第一列第二欄
#.iloc[0:1,:]  第一列

from scipy.stats.stats import pearsonr
v_1 = df.iloc[:,1]    #取第一欄
v_2 = df.iloc[:,2]    #取第二欄
p = pearsonr(v_1,v_2)
p
```

```python
#建立兩張tags*tags,0.0001的dataframe
pm = pd.DataFrame([list([0.0001]*len(tags))]*len(tags),index=tags,columns=tags)    
pv = pd.DataFrame([list([0.0001]*len(tags))]*len(tags),index=tags,columns=tags)
```

```python
for i in range(0,len(df.columns)):
    i_name = df.columns[i]
    i_list = df.iloc[:,i]
    for j in range(0,len(df.columns)):
        j_name = df.columns[j]
        j_list = df.iloc[:,j]
        pear = pearsonr(i_list,j_list) #pearson分析兩兩tag
        pm_value = round(pear[0],4)    #pm為pearson係數
        pv_value = round(pear[1],4)    #pv為p-value係數
        pm.set_value(i_name,j_name,pm_value)
        pv.set_value(i_name,j_name,pv_value)
```
