```python
with open('time_series_change.csv','r',encoding = 'cp950') as f:
    k = f.readlines()
    z = k[0].split(',')
    if z[0] == 'tag':
        z.remove('tag')
        for i in range(0,len(z)) :
            with open('tag.csv','a',newline='') as kk:        
                kk.write(z[i]+'\r\n')
                kk.close
    count = 0
    with open('taipei_weather.csv','r',encoding = 'utf-8') as ww:
        w = ww.readlines()
        for i in z:
            if z[count] == i:
                print(i)
                o = i.split('\n')[0]
                count += 1
                with open('{}.csv'.format(o),'w',newline='') as ll:
                    final = 'date'+','+'sale'+','+'ring'+','+'Stemp'+'\n'
                    ll.write(final)
                    for a in k:
                        d=a.split(',')
                        s = d[count].split('\n')[0]
                        for m in w:
                            sw = m.split(',')    
                            g = d[0].split('-')
                            t = sw[0].split('/')
                            if g[0] == t[0] and  float(g[1]) ==  float(t[1])  and  float(g[2]) ==  float(t[2]) :
                                Stemp = 1.04*float(sw[2])+0.2*(float(sw[3])/100*6.105*math.exp(17.27*float(sw[2])/(float(sw[2])+237.7))-0.65*float(sw[4])-2.7)
                                final = d[0]+','+s+','+sw[5]+','+str(Stemp)+'\n'
                                ll.write(final)
                                 
```
